```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
import random
```

## 隨意設一個loss function

```python
def L(x):
    return 6*x**4 + 3*x**3 - 43*x**2 - 3*x + 104
x = np.linspace(-3, 3, 1000)
y = 6*x**4 + 3*x**3 - 43*x**2 - 3*x + 104
plt.figure()
plt.plot(x, y)
plt.show
```

## 使用Sympy將loss function對w微分

```python
w = Symbol('w')
loss = 6*w**4 + 3*w**3 - 43*w**2 - 3*w + 104
gw = diff(loss, w)
```

## 設置隨機起點與超參數

```python
seed = random.uniform(-3, 3)
iteration = 100
lr = 0.001
momentum = 0.9
```

## Momentum 公式
![Image of Yaktocat](https://cdn-images-1.medium.com/max/1000/1*ZoRXwJ3xXxwxXPtKPAsRHg.png)

```python
plt.figure()
plt.plot(x, y)
while(1):
    plt.scatter(seed, L(seed))
    for i in range(iteration):
        if i == 0:
            v = lr * gw.subs({w:seed})
        elif i >= 1:
            v = momentum * v + lr * gw.subs({w:seed})
        seed -= v
        plt.scatter(seed, L(seed))
    break
plt.show()
```
